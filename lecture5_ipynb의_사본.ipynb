# Handling Outlier

Remove outliers of length.

Use IQR first here.

* Assume that `Length` of outlier is greater than `Q3 + (1.5 * IQR)`, less than `Q1 - (1.5 * IQR)`

```python
attribute = 'Length'


Q3= df[attribute].quantile(.75)
Q1= df[attribute].quantile(.25)
IQR = Q3-Q1
outlier_step = 1.5*IQR

filtered_df = df[(df[attribute] <Q3 + outlier_step) & (df[attribute]>Q1 - outlier_step)]
```

Draw histogram of `Length`.

```python
f, axes = plt.subplots(2, 1, figsize=(7, 5), sharex=True)

sns.histplot(data=df, x=attribute, ax=axes[0], kde=True)
sns.histplot(data=filtered_df, x=attribute, color='Red',
             ax=axes[1], kde=True)
f.tight_layout()
```

Use Z-score in here.

```python
upper_limit = df[attribute].mean() + 2 * df[attribute].std()
lower_limit = df[attribute].mean() - 2 * df[attribute].std()

filtered_df = df[(df[attribute] < upper_limit) & (df[attribute] > lower_limit)]
```

```python
f, axes = plt.subplots(2, 1, figsize=(7, 5), sharex=True)

sns.histplot(data=df, x=attribute, ax=axes[0], kde=True)
sns.histplot(data=filtered_df, x=attribute, color='Red',
             ax=axes[1], kde=True)
f.tight_layout()
```

### 1.3 Outlier 검출하기

---

눈으로 보았을 때에는 Outlier가 누구인지 대충 알 수 있지 않나요?

하지만 우리가 눈으로 보는 것만큼 컴퓨터가 이해하지는 못하겠죠? 컴퓨터가 이해하려면 정확하고 타이트한 조건을 걸어줘야 해요.

hint를 참고하여 각 속성에 따라 Outlier를 검출하는 조건을 세우고, `해당 조건에 하나라도 충족되는 사람들을` 우선 `outlier_candidate1`에 저장하세요.


**[hint]**

1. 1.2 절에서의 그래프를 주목해주세요, 해당 그래프는 `어떤 속성`의 Outlier를 `어떤 방식`으로 검출할 지 알려주고 있어요.

2. Outlier를 검출하기 위해서, `quantile`을 사용하는 방법과 `z-score`을 사용하는 방법이 있습니다. 과연 어떤 방법이 제일 적절할까요?

- 두 방법 모두 lecture note에서 설명하고 있는 표준을 따르세요 (outlier_step = 1.5, z-score = 3)


```python
###################
df.filter(['time_spent','click_per_min','refresh_per_min','delay_time'])
outlier_candidate1 = pd.DataFrame()

Q3333= df['click_per_min'].quantile(.75)
Q1111= df['click_per_min'].quantile(.25)
IQR4 = Q3333-Q1111
outlier_step4 = (1.5)*(IQR4)


up4 = df['click_per_min'].mean() + 3 * df['click_per_min'].std()
low4 = df['click_per_min'].mean() - 3 * df['click_per_min'].std()

up2 = df['delay_time'].mean() + 3 * df['delay_time'].std()
low2 = df['delay_time'].mean() - 3 * df['delay_time'].std()

Q333= df['delay_time'].quantile(.75)
Q111= df['delay_time'].quantile(.25)
IQR3 = Q333-Q111
outlier_step3 = (1.5)*(IQR3)


up3 = df['refresh_per_min'].mean() + 3 * df['refresh_per_min'].std()
low3 = df['refresh_per_min'].mean() - 3 * df['refresh_per_min'].std()

Q33= df['refresh_per_min'].quantile(.75)
Q11= df['refresh_per_min'].quantile(.25)
IQR2 = Q33-Q11
outlier_step2 = (1.5)*(IQR2)

outlier_candidate1 = df[((df['click_per_min'] >Q3333 + outlier_step4) | (df['click_per_min']<Q1111 - outlier_step4)) |
                        (df['click_per_min'] > up4) | (df['click_per_min'] < low4) |

                      
                      (df['delay_time']<Q111 - outlier_step3)|
                      ((df['delay_time'] > up2) | (df['delay_time'] < low2)) |

                      (df['refresh_per_min'] >Q33 + outlier_step2) | 
                      
                      ((df['refresh_per_min'] > up3) | (df['refresh_per_min'] < low3))
                      ]

outlier_candidate1
```
